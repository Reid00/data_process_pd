```python
#设置索引列
df_inner.set_index('id')
```

```python
#按特定列的值排序
df_inner.sort_values(by=['age','price'],ascending=[True,False])
```

```python
#按索引列排序
df_inner.sort_index(ascending=False)
```

```python
#如果 price 列的值>3000，group 列显示 high，否则显示 low
df_inner['groupped']=np.where(df_inner['price']>3000,'high','low')
df_inner
```

```python
#对复合多个条件的数据进行分组标记
df_inner.loc[(df_inner['city']=='beijing')&(df_inner['price']>=4000),'sign']=1
df_inner
```

```python
#对 category 字段的值依次进行分列，并创建数据表，索引值为 df_inner 的索引列，列名称为 category 和 size
split=pd.DataFrame((x.split('-') for x in df_inner['category']),index=df_inner.index,columns=['category','size'])
split
```

```python
#将完成分列后的数据表与原 df_inner 数据表进行匹配
df_inner=pd.merge(df_inner,split,right_index=True,left_index=True)
df_inner
```

```python
# 按条件筛选（与，或，非）
#使用“与”条件进行筛选
df_inner.loc[(df_inner['age'] > 25) & (df_inner['city'] == 'beijing'), ['id','city','age','category_x','gender']]
```

```python
#使用“或”条件筛选
df_inner.loc[(df_inner['age'] > 25) | (df_inner['city'] == 'beijing'), ['id','city','age','category_x','gender']]
```

```python
df_inner.loc[(df_inner['age'] > 25) | (df_inner['city'] == 'beijing'), ['id','city','age','category_x','gender']].sort_values(['age'])
```

```python
#对筛选后的数据按 price 字段进行求和,相当于 excel 中 sumifs 的功能

df_inner.loc[(df_inner['age'] > 25) | (df_inner['city'] == 'beijing'),['id','city','age','category_x','gender','price']].sort_values(['age']).price.sum()
```

```python
#使用“非”条件进行筛选
df_inner.loc[(df_inner['city'] != 'beijing'), ['id','city','age','category_x','gender']].sort_index()
```

```python
#对筛选后的数据按 city 列进行计数. 相当于 excel 中的 countifs 函数的功能。
df_inner.loc[(df_inner['city'] != 'beijing'), ['id','city','age','category_x','gender']].sort_values(['id']).city.count()
```

```python
#使用 query 函数进行筛选
df_inner.query('city == ["beijing", "shanghai"]')
```

```python
#对筛选后的结果按 price 进行求和
df_inner.query('city == ["beijing", "shanghai"]').price.sum()
```
